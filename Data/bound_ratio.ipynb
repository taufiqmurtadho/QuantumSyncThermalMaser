```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nhnc_ratio = np.loadtxt('smax_gap0.05.txt')[:,0]
smax_small_gap = np.loadtxt('smax_gap0.05.txt')[:,1]
smax_large_gap = np.loadtxt('smax_gap0.2.txt')[:,1]
power_small_gap = np.loadtxt('power_gap0.05.txt')[:,1]
power_large_gap = np.loadtxt('power_gap0.2.txt')[:,1]
```

```python
ratio_small_gap = np.divide(power_small_gap, smax_small_gap)
ratio_large_gap = np.divide(power_large_gap, smax_large_gap)
```

```python
plt.scatter(nhnc_ratio, ratio_small_gap)
plt.scatter(nhnc_ratio, ratio_large_gap)
```

```python
with open('bound_ratio_gap0.05.txt', 'w') as f:
    for i in range(len(nhnc_ratio)):
        f.write(str(nhnc_ratio[i])+' '+str(ratio_small_gap[i]))
        f.write('\n')
```

```python
with open('bound_ratio_gap0.2.txt', 'w') as f:
    for i in range(len(nhnc_ratio)):
        f.write(str(nhnc_ratio[i])+' '+str(ratio_large_gap[i]))
        f.write('\n')
```
