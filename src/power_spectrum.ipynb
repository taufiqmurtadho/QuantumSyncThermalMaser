```python
import numpy as np
import matplotlib.pyplot as plt
from N_Level_Engine import*
```

## PRL ($\Delta = 0)$

### N = 2

```python
#Define the parameters
sys_size = 4
omega1 = 1
omega2 = 3
gammaH = 0.05
gammaC = 0.2
nc = 0.5
hc_ratio_eng = 2
hc_ratio_ref = 0.5
nh2_ref = nc*hc_ratio_ref
nh2_eng = nc*hc_ratio_eng
Th_eng = omega2/(np.log(1+1/nh2_eng))
Th_ref = omega2/(np.log(1+1/nh2_ref))
Tc = omega1/(np.log(1+1/nc))
#p_val = 0
#p_matrix = np.array([[1,p_val],[p_val, 1]])
p_matrix = np.identity(sys_size-2)
detuning = np.linspace(-1,1,51)
gap = 0
dstrength = 0.1
```

```python
ss_list_eng = [getSteadyState(sys_size, omega1, omega2, gammaH, gammaC, Th_eng, Tc, p_matrix, gap, det, dstrength) for det in detuning]
ss_list_ref = [getSteadyState(sys_size, omega1, omega2, gammaH, gammaC, Th_ref, Tc, p_matrix, gap, det, dstrength) for det in detuning]
```

```python
power_list_eng = [powerSS(ss, omega1, omega2, gap, dstrength) for ss in ss_list_eng]
power_list_ref = [abs(powerSS(ss, omega1, omega2, gap, dstrength)) for ss in ss_list_ref]
```

```python
plt.plot(omega1/(omega2- omega1)*detuning, power_list_eng, 'r.-')
plt.plot(omega1/(omega2- omega1)*detuning, power_list_ref,'bx-')
plt.ylabel(r'$|P^{ss}|$',fontsize = 16)
plt.xlabel(r'$\delta/(\omega_2 - \omega_1)$', fontsize = 16)
plt.axvline(x = 0, linestyle = '--', color = 'black')
plt.title(r'$\Delta = 0, N = 2$', fontsize = 16)
plt.savefig('power_spectrum_N4_PRL.png', dpi = 300)
```

## PRA ($\Delta \neq 0$)

```python
#Define the parameters
sys_size = 4
omega1 = 1
omega2 = 3
gammaH = 0.1
gammaC = 0.1
nc = 0.1
hc_ratio_eng = 5
hc_ratio_ref = 0.5
nh2_eng = nc*hc_ratio_eng
nh2_ref = nc*hc_ratio_ref
Th_eng = omega2/(np.log(1+1/nh2_eng))
Th_ref = omega2/(np.log(1+1/nh2_ref))
Tc = omega1/(np.log(1+1/nc))
p_val = 0.8
p_matrix = np.array([[1,p_val],[p_val, 1]])
gap1 = 0.05
gap2 = 0.2
detuning = np.linspace(-1,1,51)
dstrength = 0.05
```

```python
ss_list_eng_small_gap = [getSteadyState(sys_size, omega1, omega2, gammaH, gammaC, Th_eng, Tc, p_matrix, gap1, det, dstrength) for det in detuning]
ss_list_eng_large_gap = [getSteadyState(sys_size, omega1, omega2, gammaH, gammaC, Th_eng, Tc, p_matrix, gap2, det, dstrength) for det in detuning]
ss_list_ref_small_gap = [getSteadyState(sys_size, omega1, omega2, gammaH, gammaC, Th_ref, Tc, p_matrix, gap1, det, dstrength) for det in detuning]
ss_list_ref_large_gap = [getSteadyState(sys_size, omega1, omega2, gammaH, gammaC, Th_ref, Tc, p_matrix, gap2, det, dstrength) for det in detuning]
```

```python
power_list_eng_small_gap = [powerSS(ss, omega1, omega2, gap1, dstrength) for ss in ss_list_eng_small_gap]
power_list_eng_large_gap = [powerSS(ss, omega1, omega2, gap2, dstrength) for ss in ss_list_eng_large_gap]
power_list_ref_small_gap = [abs(powerSS(ss, omega1, omega2, gap1, dstrength)) for ss in ss_list_ref_small_gap]
power_list_ref_large_gap = [abs(powerSS(ss, omega1, omega2, gap2, dstrength)) for ss in ss_list_ref_large_gap]
```

```python
plt.plot(omega1/(omega2- omega1)*detuning, power_list_eng_small_gap, 'r.-')
plt.plot(omega1/(omega2- omega1)*detuning, power_list_eng_large_gap, 'ro-')
plt.plot(omega1/(omega2- omega1)*detuning, power_list_ref_small_gap,'bx-')
plt.plot(omega1/(omega2- omega1)*detuning, power_list_ref_large_gap,'bv--')
plt.ylabel(r'$|P^{ss}|$',fontsize = 16)
plt.xlabel(r'$\delta/(\omega_2 - \omega_1)$', fontsize = 16)
plt.title(r'$\Delta = 0.05, N = 2$', fontsize = 16)
plt.axvline(x = 0, linestyle = '--', color = 'black')
#plt.savefig('power_spectrum_gap0.05_PRA.png')
```
